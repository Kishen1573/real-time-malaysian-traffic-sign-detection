```python
# Step 2: List your test images
val_images = [
    "/kaggle/input/malaysia-traffic-sign-yolov8-dataset/test/images/P1840082_jpg.rf.a899fa5b1a0295282da40a53bbec8894.jpg",
    "/kaggle/input/malaysia-traffic-sign-yolov8-dataset/test/images/P1840116_jpg.rf.6e81c2452344ab5f1632b46f9725d51f.jpg",
    "/kaggle/input/malaysia-traffic-sign-yolov8-dataset/test/images/P1840039_jpg.rf.4da7178eb8d972ebc06a9cb8aaa02e3b.jpg"
]
```

```python
# Step 3: Load your trained model
model_path = "/kaggle/working/yolov8s_traffic_sign/weights/best.pt"  # Adjust if needed
model = YOLO(model_path)
```

```python
# Step 4: Loop through each test image (normal color, no flip, no grayscale)
for image_path in val_images:
    # Load the image normally
    img = cv2.imread(image_path)

    # Run YOLOv8 prediction on the original color image
    results = model.predict(
        source=img,       # Directly pass color image
        conf=0.25,
        device=0          # GPU if available
    )

    # Plot results (drawn boxes)
    drawn_image = results[0].plot()

    # Convert from BGR to RGB for matplotlib display
    drawn_image = cv2.cvtColor(drawn_image, cv2.COLOR_BGR2RGB)

    # Display the image
    plt.figure(figsize=(8, 6))
    plt.imshow(drawn_image)
    plt.title(f"Detected Traffic Signs for {image_path.split('/')[-1]}")
    plt.axis("off")
    plt.show()
```

```python
# Load your results.csv
results_csv_path = '/kaggle/working/yolov8s_traffic_sign/results.csv'
df = pd.read_csv(results_csv_path)

# Show all available columns
print(df.columns.tolist())
```

```python
# Plot Loss and mAP Curves
plt.figure(figsize=(12,6))
plt.plot(df['epoch'], df['train/box_loss'], label='Box Loss (Train)')
plt.plot(df['epoch'], df['train/cls_loss'], label='Class Loss (Train)')
plt.plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP50')
plt.plot(df['epoch'], df['metrics/mAP50-95(B)'], label='mAP50-95')
plt.xlabel('Epoch')
plt.ylabel('Loss / mAP')
plt.title('YOLOv8 Training Progress - Loss and mAP')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Precision and Recall
plt.figure(figsize=(12,6))
plt.plot(df['epoch'], df['metrics/precision(B)'], label='Precision')
plt.plot(df['epoch'], df['metrics/recall(B)'], label='Recall')
plt.xlabel('Epoch')
plt.ylabel('Precision / Recall')
plt.title('YOLOv8 Precision and Recall Curve')
plt.legend()
plt.grid(True)
plt.show()
```
